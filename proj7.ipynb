```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

marks = [208,422,389,354,179,480,398]
students = ['John','James','Maybel','Mendy','Faith','Jack','Mary']
plt.pie(marks, labels=students, wedgeprops={'edgecolor':'black'},startangle=180,autopct='%1.1f%%')
```

```python
plt.close()
plt.bar(students,marks)
plt.title("Student Performances")
plt.show()
```

```python
plt.close()
plt.plot(students,marks,marker='o',ms=15)
plt.title("Student Performances")
plt.show()
```

```python
plt.close()
plt.scatter(students,marks)
plt.title("Student Performances")
plt.show()
```
