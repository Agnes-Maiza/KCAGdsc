```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#plotting data from an existingdata set
df = sns.load_dataset('iris')
df.head()

```

```python
sns.scatterplot(x='sepal_length', y ='petal_length' ,
data = df , hue = 'species')
```

```python
sns.displot(x='sepal_length', y ='petal_length' ,
data = df , hue = 'species')
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(x='sex', hue = 'survived', data = df,
palette = 'Set1')
```

```python
sns.kdeplot(x = 'age' , data = df , color = 'red')
```
